```python
# RMSE scoring metric
def rmse(y_true, y_pred):
    t = np.array(y_true)
    p = np.array(y_pred)
    return np.sqrt(mean_squared_error(t, p))

rmse = make_scorer(rmse, greater_is_better=True)
```

## Model Iterations

Finding the best regressor for the task.

##### function to search for best model params

```python
def best_model(estimator, params, X, y):
    '''
    Returns a fitted model from a GridSearchCV with the best parameters for the fitted train data.
    
    '''
    search = GridSearchCV(
                estimator,
                params, 
                scoring = 'r2',
                cv = 5,
                # random_state=42,
                n_jobs=-1, # parallelize computation
                error_score = 'raise') 
    
    search.fit(X, y)
    
    best_estimator_info = search.best_estimator_
    estimator_score = search.best_score_

    model_info = {estimator: [best_estimator_info], "estimator_score": estimator_score}
    
    return model_info
```

## Final Pipepline/s

#### Helper Functions

```python
def predict_df(pipeline: Pipeline, X_test: pd.DataFrame, y: pd.DataFrame):
    '''
    This function returns the predicted values  of the model pipes in the form of a pd.DataFrame.
    
    -----------------(PARAMETERS)-----------------------
    
    pipeline : A pipeline that was fitted on the train data.
    X_test : New X features for prediction.
    y : True y values.
    '''
    predictions = pipeline.predict(X_test)
    target_cols = y.columns
    test_preds = pd.DataFrame(predictions, columns=target_cols)
    test_preds = test_preds.round().astype(np.int32)
    
    return test_preds
```

```python
def plot_regplots(y_true, y_pred, legend_label=None):
    """
    Creates a grid of regression plots for each column in y_true and y_pred, with different colors for true and predicted values.

    Parameters:
    y_true (pandas.DataFrame): A DataFrame containing the true values.
    y_pred (pandas.DataFrame): A DataFrame containing the predicted values.
    legend_label (str): The label for the legend.
    """

    # get the number of columns in y_true
    n_cols = y_true.shape[1]

    # 2D grid of subplots
    fig, axs = plt.subplots(nrows=n_cols, ncols=1, figsize=(10, 10))

    # Iterate through each column in the df and plot the regression for each target feature
    for i in range(n_cols):
        column = y_true.columns[i]
        scatter_kws = {'facecolors':'green', 'edgecolors':'black', 'alpha':0.8}
        sns.regplot(x=y_true[column], y=y_pred[column], scatter=True, x_jitter=0.2, 
                   scatter_kws={'label': legend_label, **scatter_kws}, 
                   line_kws={'color': 'red'}, 
                   ax=axs[i])
        axs[i].set_xlabel('True Values')
        axs[i].set_ylabel('Predicted Values')
        axs[i].set_title(f'{column} : True vs Predicted Values')

    # Remove extra spines and add a legend
    sns.despine()
    handles, labels = axs[0].get_legend_handles_labels()
    if legend_label:
        handles[1].set_color('green')
    fig.legend(handles, labels, loc='upper left')
    fig.tight_layout()

    plt.show()
```

```python
# check the model's overall performance 
train_sizes_tree, train_scores_tree, test_scores_tree = learning_curve(pipe_tree, X_preproc, y, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)
```

```python
# Calculate mean and standard deviation of training and test scores
train_scores_mean_tree = np.mean(train_scores_tree, axis=1)
train_scores_std_tree = np.std(train_scores_tree, axis=1)
test_scores_mean_tree = np.mean(test_scores_tree, axis=1)
test_scores_std_tree = np.std(test_scores_tree, axis=1)

# Plot learning curve
plt.figure(figsize=(10, 6))
plt.fill_between(train_sizes_tree, train_scores_mean_tree - train_scores_std_tree,
                 train_scores_mean_tree + train_scores_std_tree, alpha=0.1,
                 color="green")
plt.fill_between(train_sizes_tree, test_scores_mean_tree - test_scores_std_tree,
                 test_scores_mean_tree + test_scores_std_tree, alpha=0.1, color="red")
plt.plot(train_sizes_tree, train_scores_mean_tree, 'o-', color="green", label="Training score")
plt.plot(train_sizes_tree, test_scores_mean_tree, 'o-', color="red", label="Test score")
plt.xlabel("Training examples")
plt.ylabel("Score")
plt.title("Tree Regressor Pipe Learning Curve")
plt.legend(loc="best")
plt.grid()
plt.show()
```

**Insight:** The model has a little bit of overfitting but not that much. As seen on the learning curve, both the train and the test have not plateaued yet, which means the performance can still be improved by training the model with more data. However, decision trees have high variance overall so this may or may not overfit if trained on more data. 

The scoring used for the model was **r2** to have a general goodness of fit for the multioutput regression task.

## Predictions

**Findings**

---
The best models to use for the tasks are DecisionTreeeRegressor and RandomForestRegressor. Both models have a slight overfitting tendency but the bias is not strong. However, each has their pros and cons. 

#### Decision Tree Regressor (pipe_tree):

- **Potential for Better Predictive Scores**: The model's train and test scores have not yet plateaued, indicating that there's potential for further improvement in predictive performance.
- **Potential for Improvement with More Data**: Since the model's scores haven't plateaued, providing more data might enhance its performance further, but it might also make it worse due to the high variance of decision trees.
- **Dispersed Residuals**: The residuals are more dispersed, suggesting that the model might capture a broader range of patterns in the data which might also miss capturing some specific data points accurately.

#### Random Forest Regressor (pipe_forest):

- **High Performance Metric ('r2')**: The model exhibits a high score for the performance metric 'r', indicating good overall performance on the test dataset.
- **Less Dispersed Residuals**: Residuals are less dispersed from the y-values, suggesting that the model is capturing the data's patterns more closely, generalizing much better.
- Although the test scores are high, they have plateaued, while the train score seems to be decreasing. 

**Conclusion**

---
Given the emphasis on accurate size predictions, I will use the Random Forest Regressor for size prediction because the **high score 'r2' score (0.86) and less dispersed residuals** suggests that it is capturing the data's patterns more closely and performs well in predicting sizes. Despite the test score plateauing, the chances of overfitting when trained on a new dataset is comparatively lower than the decision tree regressor which would give the model consistency with the predictions.

#### Tree Regressor

```python
#additional pipeline to check impact of bmi feature
pipe_additional_bmi = Pipeline([
    ('preprocessing', preproc_pipe),
    ('model', tuned_tree)
])
```

```python
pipe_bmi_score = cross_val_score(pipe_additional_bmi, X_bmi, y, cv=5, scoring='r2', error_score='raise').mean()
pipe_bmi_score
```

```python
# check the model's overall performance 
train_sizes_bmi_tree, train_scores_bmi_tree, test_scores_bmi_tree = learning_curve(pipe_additional_bmi, X_bmi_preproc, y, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)
```

```python
# Calculate mean and standard deviation of training and test scores
train_scores_mean_bmi_tree = np.mean(train_scores_bmi_tree, axis=1)
train_scores_std_bmi_tree = np.std(train_scores_bmi_tree, axis=1)
test_scores_mean_bmi_tree = np.mean(test_scores_bmi_tree, axis=1)
test_scores_std_bmi_tree = np.std(test_scores_bmi_tree, axis=1)

# Plot learning curve
plt.figure(figsize=(10, 6))
plt.fill_between(train_sizes_bmi_tree, train_scores_mean_bmi_tree - train_scores_std_bmi_tree,
                 train_scores_mean_bmi_tree + train_scores_std_bmi_tree, alpha=0.1,
                 color="green")
plt.fill_between(train_sizes_bmi_tree, test_scores_mean_bmi_tree - test_scores_std_bmi_tree,
                 test_scores_mean_bmi_tree + test_scores_std_bmi_tree, alpha=0.1, color="red")
plt.plot(train_sizes_bmi_tree, train_scores_mean_bmi_tree, 'o-', color="green", label="Training score")
plt.plot(train_sizes_bmi_tree, test_scores_mean_bmi_tree, 'o-', color="red", label="Test score")
plt.xlabel("Training examples")
plt.ylabel("Score")
plt.title("Learning Curve: Tree Regressor Pipe Trained On Dataset With BMI Feature")
plt.legend(loc="best")
plt.grid()
plt.show()
```

**Insights:** The model performance is not that different with the additional bmi feature added to the data. The test score is just a little lower compared to the model trained with the base train data.
